```python
import matplotlib.pyplot as plt

# Classifier names
classifiers = ['Logistic Regression', 'KNN', 'Random Forest', 'Naive Bayes']

# Corresponding accuracy values
accuracy_values = [0.8008095238095239, 0.9296666666666666, 0.9104285714285715, 0.7723333333333333]

# Create a bar graph
plt.bar(classifiers, accuracy_values, color='skyblue')

# Add labels and title
plt.xlabel('Classifier')
plt.ylabel('Accuracy')
plt.title('Classifier Accuracy Comparison')
plt.ylim(0.7, 1.0) 

# Display the plot
plt.show()
```
